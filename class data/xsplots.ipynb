```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = [100, 500, 1000, 2000, 6000]
U235t = [63.67, 66.24, 67.45, 67.56, 67.55]
U235f = [52.37, 56.14, 56.65, 56.8, 56.84]
U235g = [11.3, 11.1, 10.8, 10.76, 10.71]
U238t = [0.931, 0.874, 0.762, 0.751, 0.757]
U238f = [0.105, 0.11, 0.111, 0.112, 0.112]
U238g = [0.826, 0.764, 0.651, 0.639, 0.635]
```

```python
plt.figure('U235')
plt.plot(t, U235t, 'k--', label = "Total")
plt.plot(t, U235f, 'k-+', label = "Fission")
plt.plot(t, U235g, 'k-^', label = "ngamma")
plt.xlabel('Time (Days at 3E14 neutrons/s)')
plt.ylabel('Cross Section (barnes)')
plt.legend()
plt.show()
```
