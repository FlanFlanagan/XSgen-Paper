```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = [100, 500, 1000, 2000, 6000]
U235t = [63.67, 66.24, 67.45, 67.56, 67.55]
U235f = [52.37, 56.14, 56.65, 56.8, 56.84]
U235g = [11.3, 11.1, 10.8, 10.76, 10.71]
U238t = [0.931, 0.874, 0.762, 0.751, 0.757]
U238f = [0.105, 0.11, 0.111, 0.112, 0.112]
U238g = [0.826, 0.764, 0.651, 0.639, 0.635]
```

```python
plt.figure('U235')
plt.plot(t, U235t, 'k--', label = "Total")
plt.plot(t, U235f, 'k-+', label = "Fission")
plt.plot(t, U235g, 'k-^', label = r"$(n, \gamma$)")
plt.ylim([0, 70])
plt.xlabel('Time [Days]')
plt.ylabel('U235 Cross Section [b]')
plt.legend()
plt.savefig('U235xs.eps', format="eps")
plt.show()
```

```python
plt.figure('U238')
plt.plot(t, U238t, 'k--', label = "Total")
plt.plot(t, U238f, 'k-+', label = "Fission")
plt.plot(t, U238g, 'k-^', label = r"$(n, \gamma$)")
plt.ylim(0, 1.0)
plt.xlabel('Time [Days]')
plt.ylabel('U238 Cross Section [b]')
plt.legend()
plt.savefig('U238xs.eps', format="eps")
plt.show()
```
